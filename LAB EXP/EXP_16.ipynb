```python
import cv2
from google.colab.patches import cv2_imshow

# Load input image
image_path = "sample 11.png"   # Make sure file name is correct
img = cv2.imread(image_path)

if img is None:
    print("Error: Image not found. Please upload sample11.png")
else:
    print("INPUT IMAGE:")
    cv2_imshow(img)

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply Canny Edge Detection
    edges = cv2.Canny(gray, 100, 200)

    print("OUTPUT IMAGE (Canny Edges):")
    cv2_imshow(edges)
```
