```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("sample 9.jpg")

pts1 = np.float32([[10,10],[200,10],[10,200],[200,200]])
pts2 = np.float32([[20,30],[180,20],[30,220],[210,210]])

M = cv2.getPerspectiveTransform(pts1, pts2)
out = cv2.warpPerspective(img, M, (300,300))

cv2_imshow(out)
```
