```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import cv2_imshow

img = cv2.imread("sample.jpg")

if img is None:
    print("Error: Could not load image 'sample.jpg'. Please ensure the file exists and the path is correct.")
else:
    h, w = img.shape[:2]

    src = np.float32([[0,0],[w,0],[0,h],[w,h]])
    dst = np.float32([[100,50],[w-100,50],[50,h-50],[w-50,h-100]])

    H, _ = cv2.findHomography(src, dst)
    result = cv2.warpPerspective(img, H, (w, h))

    cv2_imshow(result) # Changed cv2.imshow to cv2_imshow
    # cv2.waitKey(0) and cv2.destroyAllWindows() are not needed with cv2_imshow
```
