```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import cv2_imshow for Colab

img = cv2.imread("sample 11.png") # Corrected filename from .jpg to .png
if img is None:
    print("Image not found. Please ensure 'sample 11.png' exists in the current directory.")
else:
    rows, cols = img.shape[:2]

    pts1 = np.float32([[50,50],[200,50],[50,200]])
    pts2 = np.float32([[10,100],[200,50],[100,250]])

    M = cv2.getAffineTransform(pts1, pts2)
    result = cv2.warpAffine(img, M, (cols, rows))

    print("Original Image:")
    cv2_imshow(img)
    print("Affine Transformed Image:")
    cv2_imshow(result)
    # cv2.waitKey(0) and cv2.destroyAllWindows() are not needed with cv2_imshow in Colab
```
