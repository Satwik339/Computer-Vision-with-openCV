```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

print("Experiment 21 - Closing")

img = cv2.imread("sample 6.png")

if img is None:
    print("Upload sample 6.png")
else:
    print("INPUT IMAGE:")
    cv2_imshow(img)

    kernel = np.ones((5,5), np.uint8)
    closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

    print("OUTPUT IMAGE:")
    cv2_imshow(closing)
```
